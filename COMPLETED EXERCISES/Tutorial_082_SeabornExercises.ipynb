```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Font for figures
fig_font = {'fontname':'Arial', 'size':'20'}
```

## Load in the "tips" dataset

```python
# Load the "tips" dataset 
tips = sns.load_dataset("tips")

# Take a peek at the dataset
tips.head()
```

## Add new column with tip %

```python
# Add new column with tip percentage
tips["tipPercent"] = tips.tip / tips.total_bill

# Take a look at the shiny new column
tips.head()
```

## Visualize relationship between tip %, day, and time

```python
# Visualize relationship between tip percentage, day of week, and time
sns.swarmplot(x = "tipPercent", y = "day", hue = "time", data = tips)
plt.xlabel("Tip as % of total bill", **fig_font)
plt.ylabel("Day of the week", **fig_font)
plt.show()
```

## Visualize tip %, day, time, and number of people

```python
# Use catplot to look at same data above, separated out by # people
sns.catplot(x = "tipPercent", 
            y = "day", 
            hue = "time",
            col = "size", 
            data = tips, 
            kind = "swarm",
            col_wrap = 2,
            aspect = 0.8)
plt.show()
```

## Find a trend in the "exercise" dataset

2) Find an interesting trend in the "exercise" dataset

```python
# Load the "exercise" dataset 
exercise = sns.load_dataset("exercise")

# Remove unnamed index column
exercise = exercise.drop(["Unnamed: 0"], axis = 1)

# Take a peek at the dataset
exercise.head()
```

## Let's take a look at our variables by activity

```python
# Use catplot to separate out by activity
sns.catplot(x = "time", 
            y = "pulse", 
            hue = "diet",
            col = "kind", 
            data = exercise, 
            kind = "point",
            col_wrap = 3,
            aspect = 0.5)
plt.show()
```

### Apparently there is an interaction between diet and activity such that people with no fat diets have a (substantially) higher pulse specifically when running!
